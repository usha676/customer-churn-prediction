```python
df.drop('customerID', axis=1, inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=df, x='Churn', hue='Churn', palette='Set2', legend=False)
plt.title("Distribution of Churned vs Non-Churned Customers")
plt.show()

```

```python
sns.countplot(data=df, x='gender', hue='Churn', palette='Set1')
plt.title('Churn by Gender')
plt.show()
```

```python
sns.countplot(data=df, x='Contract', hue='Churn', palette='pastel')
plt.title('Churn by Contract Type')
plt.xticks(rotation=15)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Contract', hue='Churn', palette='Set2')
plt.title("Churn by Contract Type")
plt.xlabel("Contract Type")
plt.ylabel("Number of Customers")
plt.xticks(rotation=20)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot 1
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='InternetService', hue='Churn', palette='Set2')
plt.title('Churn vs Internet Service')
plt.tight_layout()
plt.show()
```

```python
# Plot 2
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='OnlineSecurity', hue='Churn', palette='Set2')
plt.title('Churn vs Online Security')
plt.tight_layout()
plt.show()
```

```python
# Plot 3
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='TechSupport', hue='Churn', palette='Set2')
plt.title('Churn vs Tech Support')
plt.tight_layout()
plt.show()
```

```python
sns.countplot(data=df, x='Churn', palette='Set2')
plt.title('Churn Distribution')
plt.show()
```

```python
sns.countplot(data=df, x='Contract', hue='Churn', palette='Set2')
plt.title('Churn by Contract Type')
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

```python
🔍 Observation:
Customers with "Month-to-month" contracts have a much higher churn rate compared to those with "One year" or "Two year" contracts. This suggests that long-term contracts reduce churn.
```

```python
sns.countplot(data=df, x='PaymentMethod', hue='Churn', palette='Set2')
plt.title('Churn by Payment Method')
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

```python
🔍 Observation:
Customers paying via "Electronic check" have a higher tendency to churn compared to other payment methods. Automatic payments may encourage longer retention.
```

```python
plt.figure(figsize=(6,4))
sns.histplot(data=df, x='tenure', hue='Churn', multiple='stack', bins=30, palette='Set2')
plt.title('Tenure Distribution by Churn')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(data=df, x='MonthlyCharges', hue='Churn', multiple='stack', bins=30, palette='Set2')
plt.title('Monthly Charges Distribution by Churn')
plt.show()
```

```python
sns.boxplot(data=df, hue='Churn', y='MonthlyCharges', palette='Set2')
plt.title('Monthly Charges by Churn Status')
plt.show()
```

```python
sns.countplot(data=df, x='SeniorCitizen', hue='Churn', palette='Set2')
plt.title('Churn by Senior Citizen')
plt.xticks([0,1], ['No', 'Yes'])  # Optional: Rename 0 and 1
plt.show()
```

```python
## 📌 EDA Summary

- Most churn happens with "Month-to-month" contracts.
- Senior citizens are slightly more likely to churn.
- Customers with higher monthly charges and lower tenure are more prone to churn.
- Payment method also influences churn — "Electronic check" users churn more.
- Features like "Online Security", "Tech Support", and "Internet Service" show strong correlation with churn.
```
